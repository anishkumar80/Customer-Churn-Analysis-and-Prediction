```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
data = pd.read_csv("C:\\Users\\Anish\\Downloads\\Telco_Customer_Churn_Dataset .csv")
data
```

```python
data.head()
```

```python
# 1. Calculate and visualize overall churn rate

#Churn counts
churn_count = data['Churn'].value_counts()
print(churn_count)

# Calculate Churn rate
churn_rate = data["Churn"].value_counts(normalize=True) * 100
print("\n\nChurn rate (%):", churn_rate)
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Churn', data=data, palette='Set1')
plt.title('Overall Churn Distribution ')
plt.ylabel('Numbers of Customers')
plt.xlabel('Churn Status')
plt.show()
```

```python
# 2. Explore customer distributionby various demographics

#Gender Distribution 
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=data, hue='Churn', palette='pastel')
plt.title('Gender Distribution by Churn')
plt.show()
```
