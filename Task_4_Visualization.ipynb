```python
# Set style
sns.set(style="whitegrid", palette="muted")
```

```python
# 1. Box Plot:- Monthly Charges by Churn
plt.figure(figsize=(8,6))
sns.boxplot(x="Churn", y="MonthlyCharges", data=data)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
# 2. Violin Plot:- Tenure by Churn
plt.figure(figsize=(8,6))
sns.violinplot(x="Churn", y="tenure", data=data)
plt.title("Tenure vs Churn")
plt.show()
```

```python
# 3. Pair Plot:- Numerical Features
num_features = ["tenure", "MonthlyCharges", "TotalCharges"]
sns.pairplot(data[num_features + ["Churn"]], hue="Churn", diag_kind="kde")
plt.suptitle("Pair Plot of Numerical Features by Churn", y=1.02)
plt.show()
```

```python
# 4. Distribution Plot:- Total Charges by Churn
plt.figure(figsize=(8,6))
sns.histplot(data=data, x="TotalCharges", hue="Churn", kde=True, bins=30, element="step")
plt.title("Distribution of Total Charges by Churn")
plt.show()
```

```python
# 5. Box Plot - Total Charges by Contract Type and Churn
plt.figure(figsize=(10,6))
sns.boxplot(x="Contract", y="TotalCharges", hue="Churn", data=data)
plt.title("Total Charges by Contract Type (Churn vs Not Churn)")
plt.xticks(rotation=20)
plt.show()
```

```python
# 7. Stacked Bar Plots for Categorical Features
def stacked_barplot(feature, data):
    ctb = pd.crosstab(data[feature], data['Churn'], normalize='index') * 100
    ctb.plot(kind='bar', stacked=True, figsize=(8,6), colormap="Set2")
    plt.title(f"Churn Percentage by {feature}")
    plt.ylabel("Percentage")
    plt.xticks(rotation=20)
    plt.legend(title="Churn", loc="upper right")
    plt.show()
```

```python
categorical_features = ["Contract", "InternetService", "PaymentMethod", "gender", "Partner", "Dependents"]
for feature in categorical_features:
    stacked_barplot(feature,data)
```
